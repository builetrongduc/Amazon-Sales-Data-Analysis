```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/dataset/Amazon Sale Report.csv", low_memory=False)
df.info()
```

# PREPROCESSING

```python
# Loại bỏ các thuộc tính không cần thiết
df.drop(columns=['Courier Status', 'Unnamed: 22', 'ASIN', ], inplace=True)

# Xóa các dòng dữ liệu NaN 'Amount', 'ship-city'
df = df.dropna(subset=['Amount', 'ship-city'], inplace=False)

# Fill các ô có giá trị NaN ở cột 'fulfilled-by' bằng 'Amazon Ship'
df['fulfilled-by'].fillna('Amazon Ship', inplace=True)

# Fill các ô có giá trị NaN ở cột 'promotion-ids' bằng 'None'
df['promotion-ids'].fillna('None', inplace=True)

# Tách dữ liệu ở cột 'Status' ban đầu ra thành 2 cột 'Status' và 'Status-detail'
df[['Status', 'Status-detail']] = df['Status'].str.split(' - ', expand=True)

# Fill các ô có giá trị NaN ở cột 'Status-detail' bằng 'None'
df['Status-detail'].fillna('None', inplace=True)

# Đổi kiểu dữ liệu cột 'ship-postal-code' sang object
df['ship-postal-code'] = df['ship-postal-code'].astype(object)
```

```python
df.reset_index(drop=True)
```

```python
df['Date'].head()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Format the date as 'MM/DD/YYYY'
df['Date'] = df['Date'].dt.strftime('%m/%d/%Y')
```
